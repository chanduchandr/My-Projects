```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
stock_data = yf.download('TSLA', start = '2023-01-01', end = '2023-10-20')
```

```python
stock_data.head()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(stock_data['Close'], kde = True)
plt.title('TSLA Stock Closing Prices in 2023')
plt.xlabel('Date')
```

### Apply Logarithmic Transformation

```python
stock_data['LogClose'] = np.log(stock_data['Close'])
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(stock_data['LogClose'], kde=True)
plt.title('Log Transformed Distribution of Close Prices')
plt.show()
```

```python
print("Skewness of 'Close': ", stock_data['Close'].skew())
print("Skewness of 'LogClose': ", stock_data['LogClose'].skew())
```
